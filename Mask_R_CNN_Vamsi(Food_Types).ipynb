```python
annotation_file = r'D:\Food\model_detailed\send\food\labels_my-project-name_2025-02-15-12-56-27.json'
image_dir = r'D:\Food\model_detailed\send\food\img'
register_coco_instances("my_custom_train_v2", {}, annotation_file, image_dir)
```

```python
# Setup the configuration for Mask R-CNN
cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_1x.yaml"))
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_1x.yaml")

# Use your custom dataset
cfg.DATASETS.TRAIN = ("my_custom_train_v2",)
cfg.DATASETS.TEST = ()
cfg.DATALOADER.NUM_WORKERS = 4
```

```python
# Training parameters
cfg.SOLVER.IMS_PER_BATCH = 4
cfg.SOLVER.BASE_LR = 0.00025
cfg.SOLVER.MAX_ITER = 150
cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 128
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 10


# Freeze backbone layers up to the 2nd stage
cfg.MODEL.BACKBONE.FREEZE_AT = 2

# Use GPU if available
cfg.MODEL.DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# Output directory for model checkpoints
cfg.OUTPUT_DIR = './output_mask_rcnn'
os.makedirs(cfg.OUTPUT_DIR, exist_ok=True)
```

```python
import os
import torch

# Check GPU availability
print(f"CUDA Available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU Device: {torch.cuda.get_device_name(0)}")
    print(f"GPU Count: {torch.cuda.device_count()}")
else:
    print(" No GPU detected - will train on CPU (very slow)")

os.environ["CUDA_LAUNCH_BLOCKING"] = "1"
```

```python
# Start training
trainer = DefaultTrainer(cfg)
trainer.resume_or_load(resume=False)
trainer.train()
```

```python
from detectron2.engine import DefaultPredictor

# Load the trained model
cfg.MODEL.WEIGHTS = os.path.join(cfg.OUTPUT_DIR, "model_final.pth")
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5
predictor = DefaultPredictor(cfg)  # Initialize predictor
```

```python
import os
import cv2
import torch
from google.colab import files
from detectron2.utils.visualizer import Visualizer
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# Upload an image
uploaded = files.upload()  # User uploads an image
img_path = list(uploaded.keys())[0]  # Get the uploaded image filename

# Load the image
image = cv2.imread(img_path)

# Check if the image is loaded successfully
if image is None:
    print(f" Error: Failed to load image {img_path}")
else:
    print(f" Image {img_path} loaded successfully. Running inference...")

    # Perform inference
    outputs = predictor(image)

    # Get bounding box coordinates
    pred_boxes = outputs["instances"].pred_boxes.tensor.cpu().numpy()
    pred_classes = outputs["instances"].pred_classes.cpu().numpy()  # Class labels
    scores = outputs["instances"].scores.cpu().numpy()  # Confidence scores

    # Print bounding box coordinates
    print(f"\n Detected Objects in {img_path}:")
    if len(pred_boxes) == 0:
        print(" No objects detected.")
    else:
        for i, (box, cls, score) in enumerate(zip(pred_boxes, pred_classes, scores)):
            x1, y1, x2, y2 = box  # Extract bounding box coordinates
            label = metadata.thing_classes[cls]  # Get class name
            print(f" Object {i+1}: {label} | Confidence: {score:.2f} | Bounding Box -> ({x1:.2f}, {y1:.2f}, {x2:.2f}, {y2:.2f})")

    # Visualize the results
    v = Visualizer(image[:, :, ::-1], metadata=metadata, scale=1.0)
    out = v.draw_instance_predictions(outputs["instances"].to("cpu"))
    output_img = out.get_image()[:, :, ::-1]

    # Display the result
    plt.figure(figsize=(10, 6))
    plt.imshow(output_img)
    plt.axis("off")
    plt.show()
```

```python
from detectron2.evaluation import COCOEvaluator, inference_on_dataset
from detectron2.data import build_detection_test_loader

# Register the test dataset (if not already done)
register_coco_instances("my_custom_test", {}, annotation_file, image_dir)

# Update config for testing
cfg.DATASETS.TEST = ("my_custom_test",)
cfg.MODEL.WEIGHTS = os.path.join(cfg.OUTPUT_DIR, "model_final.pth")
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5  # Set confidence threshold

# Create evaluator
evaluator = COCOEvaluator("my_custom_test", cfg, False, output_dir="./output/")
test_loader = build_detection_test_loader(cfg, "my_custom_test")

# Perform evaluation
results = inference_on_dataset(predictor.model, test_loader, evaluator)
print(results)
```

```python
import shutil
import json
import os
import yaml

# Define paths
output_dir = './output_mask_rcnn'
zip_file = './food_type_model.zip'

# 1. Save model configuration
config_path = os.path.join(output_dir, "config.yaml")
with open(config_path, 'w') as f:
    f.write(cfg.dump())
print(f" Config saved at: {config_path}")

# 2. Save metadata (e.g., class names)
metadata_path = os.path.join(output_dir, "metadata.json")
metadata_dict = {"thing_classes": metadata.thing_classes}
with open(metadata_path, 'w') as f:
    json.dump(metadata_dict, f)
print(f" Metadata saved at: {metadata_path}")

# 3. Create requirements.txt to include dependencies
requirements_path = os.path.join(output_dir, "requirements.txt")
with open(requirements_path, 'w') as f:
    f.write("git+https://github.com/facebookresearch/detectron2.git@main\n")
    f.write("torch\n")
    f.write("torchvision\n")
    f.write("numpy\n")
    f.write("opencv-python\n")
print(f" requirements.txt saved at: {requirements_path}")

# 4. Create install.sh script for easy setup
install_script_path = os.path.join(output_dir, "install.sh")
with open(install_script_path, 'w') as f:
    f.write("#!/bin/bash\n")
    f.write("pip install -r requirements.txt\n")
print(f" install.sh script saved at: {install_script_path}")

# 5. Zip the output folder
shutil.make_archive('./food_type_model', 'zip', output_dir)
print(f" Model and dependencies zipped at: {zip_file}")

# 6. Download the zip file to local system
from google.colab import files
files.download(zip_file)
```
